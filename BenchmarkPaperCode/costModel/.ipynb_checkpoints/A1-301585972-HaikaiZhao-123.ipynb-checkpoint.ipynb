```python
# Normalize data to prevent large gradients
x_mean, x_std = np.mean(x), np.std(x)
y_mean, y_std = np.mean(y_noisy), np.std(y_noisy)
x_scaled = (x - x_mean) / x_std
y_scaled = (y_noisy - y_mean) / y_std

# Function to perform gradient descent with gradient clipping
def gradient_descent_with_tracking_clipped(x, y, initial_w, initial_b, learning_rate, epochs, clip_value=1.0):
    n = len(y)
    w, b = initial_w, initial_b
    cost_history = []

    for epoch in range(epochs):
        y_pred = w * x + b
        cost = loss_function(y, y_pred)
        cost_history.append(cost)

        # Calculate gradients
        dw = -(2/n) * np.sum(x * (y - y_pred))
        db = -(2/n) * np.sum(y - y_pred)

        # Clip gradients to prevent exploding gradients
        dw = np.clip(dw, -clip_value, clip_value)
        db = np.clip(db, -clip_value, clip_value)

        # Update parameters
        w -= learning_rate * dw
        b -= learning_rate * db

    return w, b, cost_history

# Exploring different learning rates with clipped gradients
plt.figure(figsize=(10, 6))
for lr in learning_rates:
    _, _, cost_history = gradient_descent_with_tracking_clipped(x_scaled, y_scaled, initial_w, initial_b, lr, epochs)
    plt.plot(range(epochs), cost_history, label=f"Learning Rate: {lr}")

plt.title('Impact of Learning Rate on Cost Convergence (Clipped Gradients)')
plt.xlabel('Epochs')
plt.ylabel('Cost')
plt.legend()
plt.grid()
plt.show()

# Rescale trained weights and biases back to the original scale
w_trained, b_trained = gradient_descent_with_tracking_clipped(
    x_scaled, y_scaled, initial_w, initial_b, 0.01, epochs
)[:2]
w_trained = w_trained * (y_std / x_std)
b_trained = b_trained * y_std + y_mean - w_trained * x_mean

# Print trained weights and biases
print(f"Trained w (slope): {w_trained:.2f}")
print(f"Trained b (intercept): {b_trained:.2f}")
```

```python
# Modified gradient descent to support various stopping criteria
def gradient_descent_with_stopping(x, y, initial_w, initial_b, learning_rate, max_epochs, stopping_criterion, threshold=None):
    n = len(y)
    w, b = initial_w, initial_b
    cost_history = []
    epoch = 0

    while epoch < max_epochs:
        y_pred = w * x + b
        cost = loss_function(y, y_pred)
        cost_history.append(cost)

        # Calculate gradients
        dw = -(2 / n) * np.sum(x * (y - y_pred))
        db = -(2 / n) * np.sum(y - y_pred)

        # Update parameters
        w -= learning_rate * dw
        b -= learning_rate * db

        # Check stopping criteria
        if stopping_criterion == "cost_threshold" and cost < threshold:
            break
        if stopping_criterion == "cost_change" and epoch > 0 and abs(cost_history[-1] - cost_history[-2]) < threshold:
            break
        if stopping_criterion == "gradient_norm" and np.sqrt(dw**2 + db**2) < threshold:
            break

        epoch += 1

    return w, b, cost_history, epoch

# Experiment with different stopping criteria
stopping_criteria = [
    {"type": "fixed_epochs", "value": 1000},
    {"type": "cost_threshold", "value": 12},
    {"type": "cost_change", "value": 1e-6},
    {"type": "gradient_norm", "value": 1e-3},
]
learning_rate = 0.01
initial_w, initial_b = 0, 0
max_epochs = 1000

# Run the experiment
plt.figure(figsize=(10, 6))
for criterion in stopping_criteria:
    if criterion["type"] == "fixed_epochs":
        w, b, cost_history, epochs = gradient_descent_with_stopping(
            x, y_noisy, initial_w, initial_b, learning_rate, criterion["value"], None
        )
    else:
        w, b, cost_history, epochs = gradient_descent_with_stopping(
            x, y_noisy, initial_w, initial_b, learning_rate, max_epochs, criterion["type"], criterion["value"]
        )
    plt.plot(range(len(cost_history)), cost_history, label=f"{criterion['type']} (value={criterion['value']})")
    print(f"Criterion: {criterion['type']}, Epochs: {epochs}, Final Cost: {cost_history[-1]:.4f}")

plt.title("Influence of Stopping Criteria on Cost Convergence")
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.legend()
plt.grid()
plt.show()
```

```python
# Track the trajectory of w, b, and costs during gradient descent
epochs=2000
def gradient_descent_trajectory(x, y, initial_w, initial_b, learning_rate, epochs, method="batch", batch_size=5):
    n = len(y)
    w, b = initial_w, initial_b
    trajectory = []

    for epoch in range(epochs):
        if method == "batch":
            y_pred = w * x + b
            cost = loss_function(y, y_pred)
            dw = -(2 / n) * np.sum(x * (y - y_pred))
            db = -(2 / n) * np.sum(y - y_pred)
        elif method == "stochastic":
            i = np.random.randint(n)  # Randomly pick one sample
            xi, yi = x[i], y[i]
            y_pred = w * xi + b
            cost = (yi - y_pred) ** 2
            dw = -2 * xi * (yi - y_pred)
            db = -2 * (yi - y_pred)
        elif method == "mini-batch":
            indices = np.random.choice(n, batch_size, replace=False)
            x_batch, y_batch = x[indices], y[indices]
            y_pred = w * x_batch + b
            cost = loss_function(y_batch, y_pred)
            dw = -(2 / len(x_batch)) * np.sum(x_batch * (y_batch - y_pred))
            db = -(2 / len(x_batch)) * np.sum(y_batch - y_pred)

        trajectory.append((w, b, cost))
        w -= learning_rate * dw
        b -= learning_rate * db

    return trajectory

# Run gradient descent methods and capture trajectories
batch_trajectory = gradient_descent_trajectory(x, y_noisy, initial_w, initial_b, learning_rate, epochs, "batch")
stochastic_trajectory = gradient_descent_trajectory(x, y_noisy, initial_w, initial_b, learning_rate, epochs, "stochastic")
mini_batch_trajectory = gradient_descent_trajectory(x, y_noisy, initial_w, initial_b, learning_rate, epochs, "mini-batch")

# Plot cost function landscape with trajectories
fig, axs = plt.subplots(1, 3, figsize=(18, 6), subplot_kw={'projection': '3d'})

# Plot Batch Gradient Descent trajectory
axs[0].plot_surface(W, B, costs, cmap='viridis', alpha=0.8)
for w, b, cost in batch_trajectory:
    axs[0].scatter(w, b, cost, color='r', s=10)
axs[0].set_title('Batch Gradient Descent')
axs[0].set_xlabel('w')
axs[0].set_ylabel('b')
axs[0].set_zlabel('Cost')

# Plot Stochastic Gradient Descent trajectory
axs[1].plot_surface(W, B, costs, cmap='viridis', alpha=0.8)
for w, b, cost in stochastic_trajectory:
    axs[1].scatter(w, b, cost, color='r', s=10)
axs[1].set_title('Stochastic Gradient Descent')
axs[1].set_xlabel('w')
axs[1].set_ylabel('b')
axs[1].set_zlabel('Cost')

# Plot Mini-Batch Gradient Descent trajectory
axs[2].plot_surface(W, B, costs, cmap='viridis', alpha=0.8)
for w, b, cost in mini_batch_trajectory:
    axs[2].scatter(w, b, cost, color='r', s=10)
axs[2].set_title('Mini-Batch Gradient Descent')
axs[2].set_xlabel('w')
axs[2].set_ylabel('b')
axs[2].set_zlabel('Cost')

plt.tight_layout()
plt.show()



```
